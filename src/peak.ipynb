```python
import pandas as pd
```

```python
Lorentz_Data = pd.read_csv('../data/Lorentz_Data.csv',sep=' ')
Lorentz_Data
```

```python
#グラフにプロット． axを用いて.点と棒どちらも表示
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(Lorentz_Data['T'],Lorentz_Data['LFxc'],'b.-',label='FMR')
ax.plot(Lorentz_Data['T'],Lorentz_Data['HFxc'],'r.-',label='Exchange')
# x軸とy軸にラベルを追加
ax.set_xlabel('Temperature [K]')
ax.set_ylabel('Frequency [GHz]')
# 'T'と'LFxc'の各データポイントのx値を表示
for (x, y) in zip(Lorentz_Data['T'], Lorentz_Data['HFxc']):
    ax.text(x, y, str(x))
# 凡例を表示
plt.legend()
# グラフを表示
plt.show()
```
