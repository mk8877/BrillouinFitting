```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

xdata = np.array([0, 1, 2, 3, 4, 5])
ydata = np.array([0,10,20 ,31, 42, 80])

def fit(x,a):
    return a*x
param, pov = sp.optimize.curve_fit(fit, xdata, ydata)

plt.plot(xdata, ydata, 'o', label='data')
plt.plot(xdata, fit(xdata, *param), label='fit')
```
